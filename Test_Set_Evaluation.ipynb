# Test Set Evaluation

In this section, the tuned models will make predictions on the test data.

```python
X_test
```

```python
# scaling training data
ct.fit(X_train)
X_train_transformed = ct.transform(X_train)
X_train[['AGE','BILIRUBIN','ALK PHOSPHATE','SGOT','ALBUMIN']] = X_train_transformed
```

```python
# scaling test data
X_test_transformed = ct.transform(X_test)
X_test[['AGE','BILIRUBIN','ALK PHOSPHATE','SGOT','ALBUMIN']] = X_test_transformed
```

```python
# checking the performance of the linear model with the best parameters(from the parameter tuning) on the test data
# NOTE: the chosen parameters for the linear model happen to be the default parameters.
log = LogisticRegression(random_state=42,C=1,solver='lbfgs',max_iter=1000)
log.fit(X_train,y_train)
log_predict = log.predict(X_test)
print('Test Accuracy: {}'.format(log.score(X_test,y_test)),'\n')
print('Test AUC: {}'.format(roc_auc_score(y_test, log.predict_proba(X_test)[:, 1])),'\n')
print('Test f1: {}'.format(f1_score(y_test,log_predict)),'\n')
print('Classification Report:\n{}'.format(classification_report(y_test,log_predict)),'\n')
print('Predictions: {}'.format(log_predict))
```

```python
# checking the performance of the neural network with the best parameters(from the parameter tuning) on the test data
mlp = MLPClassifier(max_iter=10000,random_state=42,activation='relu', alpha= 0.00001, batch_size= 30, hidden_layer_sizes= (100,), solver= 'adam')
mlp.fit(X_train,y_train)
mlp_predict = mlp.predict(X_test)
print('Test Accuracy: {}'.format(mlp.score(X_test,y_test)),'\n')
print('Test AUC: {}'.format(roc_auc_score(y_test, mlp.predict_proba(X_test)[:, 1])),'\n')
print('Test f1: {}'.format(f1_score(y_test,mlp_predict)),'\n')
print('Classification Report:\n{}'.format(classification_report(y_test,mlp_predict)),'\n')
print('Predictions: {}'.format(mlp_predict))
```

```python
# checking the performance of the support vector machine with the best parameters(from the parameter tuning) on the test data
svc = SVC(C=100,gamma=0.01,kernel = 'sigmoid',random_state=42)
svc.fit(X_train,y_train)
svc_predict = svc.predict(X_test)
print('Test Accuracy: {}'.format(svc.score(X_test,y_test)),'\n')
print('Test AUC: {}'.format(roc_auc_score(y_test, svc.decision_function(X_test))),'\n')
print('Test f1: {}'.format(f1_score(y_test,svc_predict)),'\n')
print('Classification Report:\n{}'.format(classification_report(y_test,svc_predict)),'\n')
print('Predictions: {}'.format(svc_predict))
```
